```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def MathFunc( x ):
    return np.cos( x ) - x
```

```python
print( MathFunc( np.pi ) )
```

```python
def MakeDomain( from__, to__, step ):
    return np.linspace( from__, to__, step )
```

```python
def Plot( x, y ):
    figure, xAxis = plt.subplots()
    line, = xAxis.plot( x, y( x ), label='Super Duper Plot!' )
    plt.show()
    #line.set_dashes( [2, 2, 10, 2] )
```

```python
Plot( MakeDomain( 0, 8, 500 ), MathFunc )
```

```python
p = np.linspace( .001, 20, 500 )
q = np.cos( p ) - p
fig, ax = plt.subplots()
line1, = ax.plot(p, q, label='plswork')
ax.grid()
plt.show()
```

```python
def FindPoint( begin, end, of, maxExecutions, currentExecutions ):
    midPoint = ( begin + end ) / np.float( 2.0 )
    if currentExecutions >= maxExecutions:
        return midPoint, MathFunc( midPoint )
    if ( of( midPoint ) * of( begin ) ) > 0:
        return FindPoint( midPoint, end, of, maxExecutions, currentExecutions + 1 )
    else:
        return FindPoint( begin, midPoint, of, maxExecutions, currentExecutions + 1 )
```

```python
FindPoint( np.float( 0 ), np.float( 10 ), MathFunc, 15, 0 )
```
