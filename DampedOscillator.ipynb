```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

#Define acc as F / m 
def Acceleration(x, v, k = 10, m = 0.5, b = 1.5 ):
    return  (-b * v - k * x) / m # TYPE IN SOMETHING HERE and get rid of <>
```

```python
def DampedOscillator( x0 = .4, v0 = 0 , k = 10, m = .5, b = 1.5, samples = 10000, maximumTime = 10, step = 10 / 10000 ): 
    xlist = np.empty(samples) # create lists of y and v values to store
    vlist = np.empty(samples)
    tlist = np.empty(samples)
    xlist[ 0 ]  = x0
    vlist[ 0 ]  = v0
    for i in range( 1, samples ):
        xlist[ i ] = xlist[ i - 1 ] + vlist[ i - 1 ] * step
        vlist[ i ] = vlist[ i - 1 ] + Acceleration( xlist[ i - 1 ], vlist[ i - 1 ], k, m, b ) * step
        tlist[i]  = tlist[i-1]  + step
    return tlist, xlist, vlist
  
   
```

```python
tmax = 10
x0 = .4
v0 = 0
samples = 10000
k = 10
m = .5
b = 1.5 #overdamped
lists = DampedOscillator( x0, v0, k, m, b, samples, tmax )
```
